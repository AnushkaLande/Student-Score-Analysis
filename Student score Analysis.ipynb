STUDENT SCORE ANALYSIS

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv('Expanded_data_with_more_features.csv')
df
```

SUMMARY STATISTICS AND INFO OF THE DATAFRAME

```python
df.describe()
df.info()
```

DELETING UNNAMED COLUMN

GENDER DISTRIBUTION

```python
df = df.drop("Unnamed: 0", axis = 1)
df
```

```python
plt.figure(figsize = (5,5))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

FROM ABOVE CHART WE HAVE ANALYSED THAT THE NO. OF FEMALES IS MORE THAN THE NO. OF MALES

IMPACT OF WEEKLY STUDY HOURS ON STUDEN'S SCORES

```python
gb2 = df.groupby("WklyStudyHours").agg({"MathScore":'mean', "ReadingScore":'mean', "WritingScore":'mean'})
gb2

sns.heatmap(gb2, annot = True)
plt.title("IMPACT OF WEEKLY STUDY HOURS ON STUDEN'S SCORES")
plt.show()
```

BY THE ABOVE CHART WE CAN CONCLUDE THAT THE STUDENT'S SCORES HAVE A GREAT IMPACT ON THEIR SCORES BY THE WEEKLY STUDY HOURS 

IMPACT OF PARENT EDUCATION ON STUDENTS SCORES 

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean', "ReadingScore":'mean', "WritingScore":'mean'})
gb
```

```python
sns.heatmap(gb, annot = True)
plt.title("Relation between Student's and Parent's Education")
plt.show()
```

FROM THE ABOVE CHART WE HAVE CONCLUDE THAT THE PARENTS EDUCATION HAVE A GOOD IMPACT ON THE STUDENTS

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean', "ReadingScore":'mean', "WritingScore":'mean'})
gb1
```

```python
sns.heatmap(gb1, annot = True)
plt.title("Relation between Student's Score and Parent's Marital Status")
plt.show()
```

MARITAL STATUS OF PARENTS HAVE NEGLIGIBLE IMPACT ON STUDENTS SCORE 

DETECTING THE DIFFICULT SUBJECT BETWEEN 3 (i.e., DETECTING OUTLIERS)

```python
sns.boxplot(data = df, x = 'MathScore')
plt.title('Math Scores')
plt.show()
```

```python
sns.boxplot(data = df, x = 'ReadingScore')
plt.title('Reading Scores')
plt.show()
```

```python
sns.boxplot(data = df, x = 'WritingScore')
plt.title('Writing Scores')
plt.show()
```

FROM THE ABOVE 3 CHARTS WE CAN CONCLUDE THAT MATH IS THE DIFFICULT SUBJECT FROM ALL THREE

DISTRIBUTION OF ETHNIC GROUPS

```python
groupA = df.loc[(df['EthnicGroup']== "group A")].count()
groupB = df.loc[(df['EthnicGroup']== "group B")].count()
groupC = df.loc[(df['EthnicGroup']== "group C")].count()
groupD = df.loc[(df['EthnicGroup']== "group D")].count()
groupE = df.loc[(df['EthnicGroup']== "group E")].count()

mlist = [groupA['EthnicGroup'], groupB['EthnicGroup'], groupC['EthnicGroup'], groupD['EthnicGroup'], groupE['EthnicGroup']]
l = ["Group A", "Group B", "Group C", "Group D", "Group E"]
print(mlist)

plt.pie(mlist, labels = l, autopct = "%1.2f%%")
plt.title('Distribution of Ethnic Groups')
plt.show()
```
